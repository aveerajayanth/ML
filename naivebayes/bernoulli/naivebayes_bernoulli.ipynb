```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import BernoulliNB
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('golf_df.csv')
df
```

```python
x=df.drop('Play',axis=1)
y=df['Play']
```

```python
#ENCODE PLAY
encodersy = {}        # dictionary to store encoders

le = LabelEncoder()
y= le.fit_transform(y)
encodersy = le
```

```python
from sklearn.preprocessing import LabelEncoder

encoders = {}        # dictionary to store encoders

for col in x.columns:
    le = LabelEncoder()
    x[col] = le.fit_transform(x[col])
    encoders[col] = le
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
```

```python
model = BernoulliNB()
model.fit(X_train, y_train)
```

```python
# Predict
y_pred = model.predict(X_test)
```

```python
print("Predictions:", y_pred)
print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
user_input = {
    "Outlook": "rainy",
    "Temperature": "cool",
    "Humidity": "normal",
    "Windy": "False"
}

# Encode using saved encoders
encoded_input = {}

for col in user_input:
    encoded_value = encoders[col].transform([user_input[col]])[0]
    encoded_input[col] = encoded_value

print(encoded_input)
```

```python
import numpy as np

user_df = pd.DataFrame([encoded_input])
prediction = model.predict(user_df)[0]

# Decode target (Play)
decoded_play = encodersy.inverse_transform([prediction])[0]

print("Encoded prediction:", prediction)
print("Actual prediction:", decoded_play)
```
