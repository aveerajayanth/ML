## Training and testing data

#### objective:
####    build machine learning model to predict iris species
####    split the data into two for training and testing
####    scikit learn shuffles the dataset and splits it  using train_test_split funtion
####    data=x ; label=y;

```python
# Split arrays or matrices into random train and test subsets.
from sklearn.model_selection import train_test_split
#x is data , y is target
x_train,x_test,y_train,y_test= train_test_split(iris_dataset['data'],iris_dataset['target'],random_state=0)
```

## Model building:k-Nearest neighbour

Definition:
    It is a classifier algorithm which predicts for a new datapoint from point in dataset which is close to that new point

Important parameter of KNeighbourClassifier is the number of neighbours

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(x_train,y_train)
```

```python
x_new=np.array([[5.9, 3.3 , 4.2, 1.5]])
print(x_new.shape)
```

```python
prediction=knn.predict(x_new)
print(f"Predicted target index:{prediction}")
print(f"Predicted target name:{iris_dataset['target_names'][prediction]}")
```

```python
# x_test  
```

```python
y_pred=knn.predict(x_test)
print(f"Test set prediction:{y_pred}")
```

```python
print(f"Test set score:{np.mean(y_pred==y_test)}")
```

```python
print(np.mean([1,1,1,1]==[1,1,0,1]))
```

```python
print(f"Test set score:{knn.score(x_test,y_test)}")
```
